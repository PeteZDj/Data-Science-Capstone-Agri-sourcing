## Data Understanding:

The dataset obtained from Kaggle consists of food price data collected in Kenya, providing valuable insights into the pricing dynamics of various food commodities across different regions and markets. The dataset contains the following columns:

**date**: The date of the price observation.

**admin1**: The first-level administrative division (e.g., province or county) in Kenya.

**admin2**: The second-level administrative division (e.g., district or sub-county) in Kenya.

**market**: The name of the market where the price was recorded.

**latitude**: The latitude coordinates of the market location.

**longitude**: The longitude coordinates of the market location.

**category**: The category of the food commodity (e.g., cereals and tubers, pulses and nuts, meat, fish, and eggs).

**commodity**: The specific name of the food commodity.

**unit**: The unit of measurement for the price (e.g., KG for kilograms, G for grams).

**priceflag**: Flag indicating the nature of the price observation (e.g., actual, aggregate).

**pricetype**: The type of price (e.g., Wholesale, Retail).

**currency**: The currency used for pricing (e.g., KES for Kenyan Shilling).

**price**: The price of the commodity in the local currency.

**usdprice**: The price of the commodity converted to USD.

The dataset contains 10,767 rows, each representing a unique observation of food prices in a specific market on a particular date. This dataset provides a comprehensive view of food pricing trends and variations across different regions and market types in Kenya, offering valuable insights for analysis and decision-making in the agricultural sector.

```python
# Standardazing our metrics 
conversion_factors = {
    'KG': 1,
    '90 KG': 90,
    '400 G': 0.4,
    '50 KG': 50,
    '500 ML': 0.5,
    'L': 1,  
    '200 ML': 0.2,
    'Unit': 1,  
    'Bunch': 1,  
    '64 KG': 64,
    '13 KG': 13,
    '126 KG': 126,
    '200 G': 0.2,
    'Head': 1  
}


def standardize_unit(row):
    unit = row['unit']
    if unit in conversion_factors:
        return conversion_factors[unit]
    else:
        return None  


df['standardized_unit'] = df.apply(standardize_unit, axis=1)

print(df.head())  
```

## Association Analysis

```python

#Select relevant columns
selected_columns = ['date', 'Provinces', 'Counties', 'commodity']

#Convert to Transactional Format
transaction_data = df[selected_columns].copy()

# Converting categorical variables to binary indicators using one-hot encoding
transaction_data = pd.get_dummies(transaction_data, columns=['commodity'], prefix='', prefix_sep='')

# Group transactions by date, provinces, and counties and aggregate commodities
transaction_data = transaction_data.groupby(['date', 'Provinces', 'Counties']).sum()

# Reset index to make date, provinces, and counties columns again
transaction_data.reset_index(inplace=True)

# Convert commodity counts to binary values (1 if bought, 0 otherwise), excluding the date column
for column in transaction_data.columns[3:]:
    transaction_data[column] = transaction_data[column].apply(lambda x: 1 if x > 0 else 0)

transaction_data.head()

# 
```

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules


 #Finding the frequent itemsets using Apriori algorithm
frequent_itemsets = apriori(transaction_data.iloc[:, 3:], min_support=0.1, use_colnames=True)

# Generating association rules
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)

# Filtering and displaying the rules based on desired metrics (support, confidence, lift)
filtered_rules = rules[(rules['support'] >= 0.1) &
                      (rules['confidence'] >= 0.5) &
                      (rules['lift'] >= 1.2)]


filtered_rules
```

```python

# Support
rules['support'] = rules['antecedent support'] * rules['consequent support']

# Confidence
rules['confidence'] = rules['support'] / rules['antecedent support']

# Lift
rules['lift'] = rules['support'] / (rules['antecedent support'] * rules['consequent support'])

# Conviction
rules['conviction'] = (1 - rules['consequent support']) / (1 - rules['confidence'])

# Leverage
rules['leverage'] = rules['support'] - (rules['antecedent support'] * rules['consequent support'])

# Interest
rules['interest'] = rules['confidence'] / rules['consequent support']

# Coverage (optional, not directly calculated from the association rules DataFrame)
total_transactions = len(transaction_data)
rules['coverage'] = rules['support'] * total_transactions

# Display the updated rules DataFrame with calculated metrics
rules.head()
```

## Market Basket Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Define small and large baskets with quantities
small_basket = { 
    'Maize': 2, 'Bread': 2, 
    'Sugar': 1,
    'Meat (beef)': 1, 'Sorghum': 2, 'Potatoes (Irish)': 10, 'Meat (camel)': 2,
    'Salt': 1, 'Fuel (kerosene)': 1, 
    'Spinach': 1, 'Tomatoes': 4, 'Onions (dry)': 2,
    'Cooking fat': 1, 'Milk (UHT)': 1
}

large_basket = { 
    'Maize': 6, 'Bread': 8, 
    'Sugar': 2,
    'Meat (beef)': 3, 'Sorghum': 4, 'Potatoes (Irish)': 30, 'Meat (camel)': 6,
    'Salt': 2, 'Fuel (kerosene)': 10, 
    'Spinach': 1, 'Tomatoes': 4, 'Onions (dry)': 10,
    'Cooking fat': 10, 'Milk (UHT)': 10
}

# Calculate total price for each year
total_prices_small = {}
total_prices_large = {}

for year in df['year'].unique():
    # Filter dataset for the specific year
    data_year = df[df['year'] == year]
    
    # Calculate total price of small basket for the year
    total_price_small = sum(data_year[data_year['commodity'] == k]['price'].values[0] * v if not data_year[data_year['commodity'] == k].empty else 0 for k, v in small_basket.items())
    total_prices_small[year] = total_price_small
    
    # Calculate total price of large basket for the year
    total_price_large = sum(data_year[data_year['commodity'] == k]['price'].values[0] * v if not data_year[data_year['commodity'] == k].empty else 0 for k, v in large_basket.items())
    total_prices_large[year] = total_price_large

# Convert total_prices dictionaries to lists for plotting
years_list = list(df['year'].unique())
small_prices_list = [total_prices_small[year] for year in years_list]
large_prices_list = [total_prices_large[year] for year in years_list]

# Create DataFrame for plotting
df_total_prices = pd.DataFrame({
    'Year': years_list * 2,
    'Basket': ['Small'] * len(years_list) + ['Large'] * len(years_list),
    'Total Price': small_prices_list + large_prices_list
})

# Plotting
plt.figure(figsize=(10, 6))
sns.lineplot(data=df_total_prices, x='Year', y='Total Price', hue='Basket', palette='muted')
plt.xlabel('Year')
plt.ylabel('Total Price')
plt.title('Total Price of Small and Large Baskets for Each Year')
plt.xticks(rotation=45)
plt.legend(title='Basket')
plt.grid(True)
plt.show()

```

- Both the total prices of small and large baskets show an increasing trend over the years. This suggests that the cost of living, as represented by the prices of these baskets, has generally been rising.

Regression model using multiple variables

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
import statsmodels.api as sm
import pandas as pd


# Split data into train and test sets
X1 = df[['date', 'Provinces', 'Counties', 'market', 'category', 'commodity', 'pricetype']]
y1 = df['price_per_standard_unit']

X_train1, X_test1, y_train1, y_test1 = train_test_split(X1, y1, test_size=0.2, random_state=42)

# Define transformers for categorical columns
categorical_transformer = OneHotEncoder()

# Column transformer to handle both types of columns
preprocessor = ColumnTransformer(
    transformers=[
        ('cat', categorical_transformer, ['date', 'Provinces', 'Counties', 'market', 'category', 'commodity', 'pricetype'])
    ],
    remainder='passthrough'
)

# Create a pipeline with preprocessing and model
pipeline = Pipeline([
    ('preprocessor', preprocessor),
    ('regressor', LinearRegression())
])

# Fit the pipeline
pipeline.fit(X_train1, y_train1)

# Get preprocessed data
X_train_preprocessed = pipeline.named_steps['preprocessor'].transform(X_train1)

# Convert sparse matrix to dense matrix
X_train_preprocessed = X_train_preprocessed.toarray()

# Add constant to the independent variables
X_train_preprocessed = sm.add_constant(X_train_preprocessed)

# Fit the OLS model
model = sm.OLS(y_train1, X_train_preprocessed).fit()

# Print the model summary
print(model.summary())
```

```python
# Calculate residuals
residuals1 = model.resid

# Predicted values
predicted_values1 = model.predict(X_train_preprocessed)

# Plot residuals vs predicted values
plt.figure(figsize=(10, 6))
sns.scatterplot(x=predicted_values1, y=residuals1)
plt.axhline(y=0, color='r', linestyle='-')
plt.title('Residuals vs Predicted Values')
plt.xlabel('Predicted Values')
plt.ylabel('Residuals')
plt.show()
```
